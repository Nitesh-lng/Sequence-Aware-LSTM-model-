```python
tokenizer=get_tokenizer('basic_english')
```

```python
def example_tokenizer(example,tokenizer,max_length):
  tokens=tokenizer(example['text'])[:max_length]
  length=len(tokens)
  return {'tokens':tokens,'length':length}
```

```python
max_length=156
```

```python
train_data=train_data.map(
    example_tokenizer,fn_kwargs={
        'tokenizer':tokenizer,
        'max_length':max_length})

test_data=test_data.map(
    example_tokenizer,fn_kwargs={
        'tokenizer':tokenizer,
        'max_length':max_length
    })
```

```python
min_freq=3
special_tokens=['<unk>','<pad>']
vocab=build_vocab_from_iterator(
    train_data['tokens'],
    min_freq=min_freq,
    specials=special_tokens
)
```

```python
def numeralization(example,vocab):
  ids=vocab.lookup_indices(example['tokens'])
  return {'ids':ids}
```

```python
train_data=train_data.map(
    numeralization,fn_kwargs={
        'vocab':vocab
    }
)

test_data=test_data.map(
    numeralization,fn_kwargs={
        'vocab':vocab
    }
)

valid_data=valid_data.map(
    numeralization,fn_kwargs={
        'vocab':vocab
    }
)
```

```python
import torch
from torch.utils.data import DataLoader
from torch.nn.utils.rnn import pad_sequence

def collate_fn(batch):
    batch_ids = [item["ids"] for item in batch]
    batch_ids = pad_sequence(batch_ids, batch_first=True, padding_value=0)

    batch_length = [len(item["ids"]) for item in batch]
    batch_length = torch.tensor(batch_length)
    batch_labels = [item["label"] for item in batch]
    batch_labels = torch.tensor(batch_labels)

    return {"ids": batch_ids, "length": batch_length, "label": batch_labels}

def get_data_loader(dataset, batch_size, shuffle=False):
    return DataLoader(dataset, batch_size=batch_size, shuffle=shuffle, collate_fn=collate_fn)
```

```python
batch_size=512
train_data_loader=get_data_loader(train_data,batch_size,pad_index)
valid_data_loader=get_data_loader(valid_data,batch_size,pad_index)
test_data_loader=get_data_loader(test_data,batch_size,pad_index)
```

```python
class LSTMModel(nn.Module):
  def __init__(self,vocab_size,embed_size,hidden_size,output_size,pad_idx,dropout=0.5):
    super(LSTMModel,self).__init__()
    self.embedding=nn.Embedding(
        vocab_size,
        embed_size,
        padding_idx=pad_idx
    )
    self.lstm=nn.LSTM(
        embed_size,
        hidden_size,
        batch_first=True,
        dropout=dropout
    )
    self.fc=nn.Linear(
        hidden_size,
        output_size
    )
    self.dropout=nn.Dropout(dropout)

  '''def forward(self,x,length):
    embedded=self.embedding(x)
    packed_embedded=nn.utils.rnn.pack_padded_sequence(
        embedded,
        length,
        batch_first=True,
        enforce_sorted=False
    )
    packed_output,(hidden,cell)=self.lstm(packed_embedded)
    hidden=self.dropout(hidden[-1])
    output=self.fc(hidden)
    return output'''


  def forward(self, x, lengths):
   embedded = self.embedding(x)

   lengths = lengths.cpu().long()

   packed_embedded = nn.utils.rnn.pack_padded_sequence(
       embedded,
       lengths,
       batch_first=True,
       enforce_sorted=False
   )

   packed_output, (hidden, cell) = self.lstm(packed_embedded)

   hidden = self.dropout(hidden[-1])

   output = self.fc(hidden)

   return output
```

```python
embed_size=128
hidden_size=256
output_size=2
learning_rate=0.001
epochs=20

model=LSTMModel(
    vocab_size=len(vocab),
    embed_size=embed_size,
    hidden_size=hidden_size,
    output_size=output_size,
    pad_idx=pad_index
)

criterion=nn.CrossEntropyLoss()

optimizer=optim.Adam(model.parameters(),lr=learning_rate)

device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

```python
def train_model(model,train_loader,valid_loader,criterion,optimizer,epochs):
  for epoch in range(epochs):
    model.train()
    train_loss=0
    correct_predictions=0
    total_predictions=0

    for batch in tqdm.tqdm(train_loader,desc=f"Epoch {epch+1}/{epochs}"):
      inputs,labels,length=batch['ids'].to(device),batch['labels'].to(device),batch['length'].to(device)
      optimizer.zero_grad()
      ouputs=model(inputs,length)
      loss=criterion(outputs,labels)
      train_loss +=loss.items()

      loss.backward()
      optimizer.step()

      _, preds = torch.max(outputs, 1)
      correct_predictions += (preds == labels).sum().item()
      total_predictions += labels.size(0)


    avg_train_loss = train_loss / len(train_loader)
    train_accuracy = correct_predictions / total_predictions

    model.eval()
    valid_loss = 0
    correct_predictions = 0
    total_predictions = 0

    with torch.no_grad():
      for batch in valid_loader:
        inputs, labels, lengths = batch['ids'].to(device), batch['label'].to(device), batch['length'].to(device)

        outputs = model(inputs, lengths)

        loss = criterion(outputs, labels)
        valid_loss += loss.item()

        _, preds = torch.max(outputs, 1)
        correct_predictions += (preds == labels).sum().item()
        total_predictions += labels.size(0)


  avg_valid_loss = valid_loss / len(valid_loader)
  valid_accuracy = correct_predictions / total_predictions


  print(f"Epoch {epoch + 1}/{epochs}")
  print(f"Train Loss: {avg_train_loss:.4f}, Train Accuracy: {train_accuracy:.4f}")
  print(f"Valid Loss: {avg_valid_loss:.4f}, Valid Accuracy: {valid_accuracy:.4f}")


  train_model(
      model,
      train_data_loader,
      valid_data_loader,
      criterion,
      optimizer,
      epochs
  )

```

```python
def evaluate_model(model, test_loader):
    model.eval()
    correct_predictions = 0
    total_predictions = 0
    test_loss = 0

    with torch.no_grad():
        for batch in test_loader:
            inputs, labels, lengths = batch['ids'].to(device), batch['label'].to(device), batch['length'].to(device)

            outputs = model(inputs, lengths)

            loss = criterion(outputs, labels)
            test_loss += loss.item()

            _, preds = torch.max(outputs, 1)
            correct_predictions += (preds == labels).sum().item()
            total_predictions += labels.size(0)

    avg_test_loss = test_loss / len(test_loader)
    test_accuracy = correct_predictions / total_predictions

    print(f"Test Loss: {avg_test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")

evaluate_model(model, test_data_loader)
```
